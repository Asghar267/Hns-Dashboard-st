```python
import pandas as pd
import numpy as np
from services.qr_commission_service import QRCommissionService
from modules.blink_reporting import prepare_blink_orders

start_date = "2026-03-01"
end_exclusive = "2026-03-11"  # end date exclusive
branches = [2]
mode = "Filtered"  # or "Unfiltered"
```

```python
svc = QRCommissionService()

non_blink = svc.get_non_blinkco_sales_transactions(start_date, end_exclusive, branches, mode)
non_blink["external_ref_id"] = non_blink["external_ref_id"].astype(str).str.strip().replace("None", "")
non_no_ref = non_blink[non_blink["external_ref_id"].eq("")].copy()

print("Non-Blinkco no-ref rows:", len(non_no_ref))

raw = svc.get_blink_raw_orders(start_date, end_exclusive)
raw = raw[["BlinkOrderId", "OrderJson", "CreatedAt"]].copy()
raw = prepare_blink_orders(raw)

raw_prices = pd.to_numeric(raw.get("Indoge_total_price", 0), errors="coerce").dropna()
raw_prices = raw_prices[raw_prices > 0].astype(float).values

print("Blink raw price rows:", len(raw_prices))

if len(non_no_ref) == 0 or len(raw_prices) == 0:
    print("No data for price-only match.")
else:
    raw_prices_sorted = np.sort(raw_prices)

    def nearest_price(x: float) -> float:
        idx = np.searchsorted(raw_prices_sorted, x, side="left")
        if idx == 0:
            return float(raw_prices_sorted[0])
        if idx >= len(raw_prices_sorted):
            return float(raw_prices_sorted[-1])
        prev_v = raw_prices_sorted[idx - 1]
        next_v = raw_prices_sorted[idx]
        return float(prev_v if abs(x - prev_v) <= abs(x - next_v) else next_v)

    non_no_ref["total_sale"] = pd.to_numeric(non_no_ref["total_sale"], errors="coerce").fillna(0.0)
    non_no_ref = non_no_ref[non_no_ref["total_sale"] > 0].copy()

    non_no_ref["closest_indoge_price"] = non_no_ref["total_sale"].apply(nearest_price)
    non_no_ref["difference"] = non_no_ref["total_sale"] - non_no_ref["closest_indoge_price"]
    non_no_ref["diff_pct"] = (non_no_ref["difference"].abs() / non_no_ref["total_sale"].replace(0, np.nan)) * 100.0
    non_no_ref["price_match_exact"] = non_no_ref["difference"].abs() < 0.01

    exact = int(non_no_ref["price_match_exact"].sum())
    match_pct = (exact / len(non_no_ref) * 100.0) if len(non_no_ref) else 0.0

    print("Exact matches:", exact)
    print("Exact match %:", round(match_pct, 2))
    print("Mean diff %:", round(non_no_ref["diff_pct"].mean(), 2))

non_no_ref[[
    "sale_id",
    "shop_id",
    "employee_id",
    "employee_code",
    "employee_name",
    "total_sale",
    "closest_indoge_price",
    "difference",
    "diff_pct",
    "price_match_exact"
]].sort_values("total_sale", ascending=False).head(20)
```
